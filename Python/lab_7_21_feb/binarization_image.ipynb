```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import cv2 as cv
img = cv.imread("D:/Programing_works/6th_semester_work/Python/lab_7_21_feb/coin.jpeg", 1)
#img = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
plt.imshow(img)
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
blurred = cv.GaussianBlur(gray, (7, 7), 0)
plt.imshow(blurred)
plt.title("Gaussian blurred")
plt.show()
# cv.imshow("Gaussian blurred", blurred)
# cv.waitKey(0)
# cv.destroyAllWindows()
```

```python
(T, thresh) = cv.threshold(blurred, 200, 255, cv.THRESH_BINARY)
plt.imshow(thresh)
plt.title("Threshold Binary")
plt.show()
# cv.imshow("Threshold Binary", thresh)
# cv.waitKey(0)
# cv.destroyAllWindows()
```

```python
(T, threshInv) = cv.threshold(blurred, 200, 255,cv.THRESH_BINARY_INV)
plt.imshow(threshInv)
plt.title("Threshold Binary Invented")
plt.show()
# cv.imshow("Threshold", threshInv)
# cv.waitKey(0)
# cv.destroyAllWindows()
```

```python

img = cv.imread("D:/Programing_works/6th_semester_work/Python/lab_7_21_feb/coin.jpeg",1)
gray = cv.cvtColor(img.copy(),cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(gray, 125, 255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
copy_img = img.copy()
final = cv.drawContours(copy_img,contours,-1,(0,0,0),2)
plt.imshow(copy_img)
```
